```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy
import h5py
from sys import path
path.append("/home/hkli/work/fourier_quad/")
from Fourier_Quad import Fourier_Quad
import tool_box
```

```python
f = h5py.File("data_0.hdf5","r")
data = f["/data"].value
detect = data[:,-1]
mag = -data[:,-2]
idx_1 = mag > 22.5
idx = detect > -1
idx = idx & idx_1
pk0 = data[:,4][idx]/64/60
pk0_fit = data[:,5][idx]/64/60
pk0_max = data[:,6][idx]/64/60
mag = mag[idx]

selects = [pk0, pk0_fit, pk0_max]
labels = ["P$_{k0}$","P$_{k0}$-fit","MAX(P$_{k0}$,P$_{k0}$-fit)"]
```

```python

def scatter_plot(x, y, mode,colors, y_lim, pic_path=None):
    fmt='%2.f%%'
    fig_x = 8
    fig_y = fig_x*4/6
    figs = (fig_x, fig_y)
    fonts = 20
    xy_lb_size = 18
    legend_size = xy_lb_size - 4
    axis_linewidth = 1.2
    plt_line_width = 2
    cap_size = 5
    xticks = mtick.FormatStrFormatter(fmt)
    fig = plt.figure(figsize=figs)

    ax = fig.add_subplot(111)
    ax.tick_params(direction='in', labelsize=xy_lb_size, top=True, right=True)
    for axis in ["bottom", "left", "top", "right"]:
        # the line width of the frame
        ax.spines[axis].set_linewidth(axis_linewidth)
    ax.xaxis.set_tick_params(which="both",direction="in",length=6, width=axis_linewidth)
    ax.yaxis.set_tick_params(which="major",direction="in",length=8, width=axis_linewidth)
    ax.yaxis.set_tick_params(which="minor",direction="in",length=4, width=axis_linewidth)

    marks = ["v", "s", "p"]
    for i in range(len(colors)):
        if mode == "scatter":
            ax.scatter(x, y[i], s=18,marker=marks[i],facecolors='none', color=colors[i],label=labels[i])
        else:
            ax.fill_between(mags, pks[i][:,0],pks[i][:,1], alpha=0.5,color=colors[i],label=labels[i])
    #     ax.plot(mag[ch], selects[i][ch], color=colors[i],label=labels[i])
#     ys = ax.set_ylim()
#     ax.set_ylim(0.05, ys[1]+50)
    ax.set_ylim(y_lim[0],y_lim[1])
    ax.set_yscale("log")
    ax.set_xlabel("Magnitude",fontsize=xy_lb_size)
    ax.set_ylabel("P$_{k0}$",fontsize=xy_lb_size)
#     ax.legend(loc="best", bbox_to_anchor=(0.98, 0.98))
    ax.legend(loc="lower left", bbox_to_anchor=(0.04, 0.04),fontsize=legend_size)
    if pic_path:
        plt.savefig(pic_path,bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
ch = numpy.random.choice(numpy.array([i for i in range(len(mag))]), 10000, replace=False).tolist()
```

```python
plot_data = [selects[i][ch] for i in range(3)]
scatter_plot(mag[ch],plot_data,"scatter",["C0","C1","C2"],[0.05,60])
```

```python
pks = [[],[],[]]
mags = []
num = 500
magc = mag[ch]
d_mag = (magc.max() - 22.5)/num
for i in range(num):
    idx1 = magc >= 22.5 + i*d_mag
    idx2 = magc< 22.5 + (i+1)*d_mag
    mags.append(22.5+(i+0.5)*d_mag)
    for j in range(3):
        data = selects[j][ch][idx1&idx2]
        pks[j].append((data.min(), data.max()))
pks = numpy.array(pks)
pks.shape
len(pks)

scatter_plot(magc, pks, "belt",["C0", "C1","C2"],[0.05, 60],"/home/hkli/work/jupyter_run/pk0_scatter.pdf")
```
