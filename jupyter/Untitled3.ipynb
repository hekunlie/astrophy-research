```python
npz = numpy.load("/home/hkli/work/test/pair_data.npz")
data = npz["arr_0"]
pair_num = data.shape[0]
print(data.shape)

for i in range(10):
    print(data[0,i])
    if i == 5:
        print("\n")
```

```python
coeff = 299792.458/100

for i in range(pair_num):
    ra_b = data[0,0]
    dec_b = data[0,1]
    cos_dec_b = data[0,2]
    dist_b = data[0,3]
    z_b = data[0,4]

    ra_f = data[0,6]
    dec_f =  data[0,7]
    dist_f =  data[0,8]
    z_f = data[0,9]
    diff_ra = (ra_b - ra_f)*cos_dec_b
    diff_dec = dec_b - dec_f
    diff_theta = numpy.sqrt(diff_ra*diff_ra + diff_dec*diff_dec)
    diff_rad = diff_theta*numpy.pi/180
    
    c1 = SkyCoord(ra=ra_f*units.deg,dec=dec_f*units.deg,frame ="fk5" )
    c2 = SkyCoord(ra=ra_b*units.deg,dec=dec_b*units.deg,frame ="fk5" )
    diff_rad_a = c1.separation(c2).radian
    
    print(dist_b*coeff*diff_rad, dist_b*coeff*diff_rad_a)
```
