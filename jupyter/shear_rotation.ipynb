```python
def rotation(e1, e2, theta):
#     e = (e1 + i e2)*exp(2i\theta)
    e1_r = e1*numpy.cos(2*theta) - e2*numpy.sin(2*theta)
    e2_r = e1*numpy.sin(2*theta) + e2*numpy.cos(2*theta)
    if numpy.abs(e1_r) < 1.e-6:
        e1_r = 0
    if numpy.abs(e2_r) < 1.e-6:
        e2_r = 0
    return e1_r, e2_r

def find_angle(e1,e2):
    e = numpy.sqrt(e1**2+e2**2)
    cos_theta = numpy.sqrt((1+e1/e)/2)
    
    if numpy.abs(e1) < 1.e-8:
        e1 = 0
    if numpy.abs(e2) < 1.e-8:
        e2 = 0
    
    if cos_theta == 0:
        sin_theta = 1
    else:
        sin_theta = e2/e/2/cos_theta
        
    if e2 < 0:
        cos_theta = -cos_theta
        sin_theta = -sin_theta
        
    return cos_theta, sin_theta
```

```python
img = plot_tool.Image_Plot()
img.create_subfig(1,1)
alpha = 1./numpy.pi*180
for i in range(8):
    theta_0 = i*numpy.pi*2/16 + 22.5/180*numpy.pi
    er = rotation(0.3,0,theta_0)
    theta = find_angle(er[0],er[1])
    img.axs[0][0].plot([theta[0],-theta[0]],[theta[1], -theta[1]], label="%4f"%(theta_0*alpha))
#     print(er)
#     print("e1: %6.3f, e2: %6.3f. "%(er[0], er[1]))
#     print("%4.1f, %6.3f, %6.3f, %6.3f, %6.3f,"%(theta_0*alpha,numpy.cos(theta_0), theta[0], numpy.sin(theta_0),  theta[1]))
#     print("\n")
img.axs[0][0].legend()
```
