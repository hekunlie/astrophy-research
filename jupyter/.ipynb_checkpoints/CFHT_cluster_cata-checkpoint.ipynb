```python
%matplotlib inline
import numpy
from astropy.io import fits
import h5py
from sys import  path
path.append("/home/hkli/work/mylib/")
import tool_box
from plot_tool import Image_Plot
import matplotlib.pyplot as plt
import os
```

```python
data_path = "/mnt/ddnfs/data_users/hkli/CFHT/gg_lensing/data/foreground/CFHT_cluster/"
```

```python
img = Image_Plot()
img.create_subfig(2,2)
for i in range(1,5):
    data = numpy.loadtxt(data_path + "Clusters_W%d.dat"%i)
    ra = data[:,0]
    dec = data[:,1]
    z = data[:,2]
    sig = data[:,3]
    n200 = data[:,4]   

    idx = sig >= 5
    h5f = h5py.File(data_path + "w_%d.hdf5"%i,"w")
    h5f["/RA"] = ra[idx]
    h5f["/DEC"] = dec[idx]
    h5f["/Z"] = z[idx]
    h5f["/COS_DEC"] = numpy.abs(numpy.cos(dec[idx]/180*numpy.pi))
    h5f.close()

    
    m,n = divmod(i-1, 2)
    
#     norm = plt.Normalize(vmin=numpy.min(z[idx]), vmax=numpy.max(z[idx]))
#     cmap = plt.get_cmap('YlOrRd')
#     cl = cmap(norm(z[idx]))
    
    norm = plt.Normalize(vmin=numpy.min(n200[idx]), vmax=numpy.max(n200[idx]))
    cmap = plt.get_cmap('YlOrRd')
    cl = cmap(norm(n200[idx]))
    
    img.axs[m][n].scatter(ra[idx], dec[idx],color=cl, s=3)
    
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm._A = []
    
    plt.colorbar(sm, ax=img.axs[m][n])
    print(data.shape, idx.sum())
```
